```python
def benchmark_results (root):
    for test in os.listdir(root):
        for fname in os.listdir(os.path.join(root, test)):
            if '.jsonl' in fname:
                continue
            path = os.path.join(os.path.join(root, test), fname)
            with open(path, 'rb') as f:
                metrics = json.load(f)
                metrics.pop('prompts')
            print("=" * 50, test, "=" * 50)
            print(metrics)
            print('\n')
```
